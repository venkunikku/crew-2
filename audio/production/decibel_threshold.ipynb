```python
import numpy as np
import scipy.io.wavfile as wavfile

import matplotlib.pyplot as plt

```

```python
noise_file = "../../../noise/combined-noise.wav"
sr, noise = wavfile.read(noise_file)
noise = noise[:44100*4]
power_db = 20 * np.log10(noise / (2**15))
power_db = power_db[~np.isnan(power_db)]
power_db = power_db[~np.isinf(power_db)]
plt.plot(power_db)
np.mean(power_db)
```

```python
# 16-bit conversion

#valueDBFS = 20*np.log10(abs(noise)/32768)
freq_signal = np.fft.fft(noise)
len_signal = len(noise)
len_half = np.ceil((len_signal + 1) / 2.0).astype(np.int) 
freq_signal = abs(freq_signal[0:len_half]) / len_signal
freq_signal **= 2  
len_fts = len(freq_signal)

if len_signal % 2:
    freq_signal[1:len_fts] *= 2
else:
    freq_signal[1:len_fts-1] *= 2
    
signal_power = 10 * np.log10(freq_signal)


# Build the X axis
x_axis = np.arange(0, len_half, 1) * (44100 / len_signal) / 1000.0

# Plot the figure
plt.figure()
plt.plot(x_axis, signal_power, color='black')
plt.xlabel('Frequency (kHz)')
plt.ylabel('Signal power (dB)')
plt.show()

np.mean(signal_power)
```
